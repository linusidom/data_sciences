### Metrics and Evaluation


###### Evaluations

Each model will have it's own method of evaluation

Regression Models rely on
- Mean Absolute Error
 - Average of the Sum of the Absolute difference between the data points and the line of best fit
 - Doesn't do well with outliers as outliers can skew the data heavily
- Mean Square Error
 - Average of the Sum of the Absolute difference squared between the data points and the line of best fit
 - Averages out the outliners so they don't have that big of an impact on the final result
- Root Mean Squared Error
 - Square Root of the Mean Squared Error so that the final value makes sense
- R2 Value aka Score
 - Percent accuracy of the line of best fit with the datapoints

The following are score guidelines, not facts
 - 70-80% considered okay but could be a sign of underfitting
 - 80-90% considered good or balanced
 - 90-100% considered great but could be a sign of overfitting and being too specific to the dataset

The results are dependent on the type of data and whether they are within acceptable error ranges

###### Coefficients

### Problems with Machine Learning

###### Overfitting 

The model was far too correct

High correlation between the model and the data

###### Underfitting

The model was too far off

Low correlation between the model and the data

###### Bias

Relates to target

Low Bias - Results are close to the intended target

High Bias - Results are far away from the intended Target

###### Variance

Relates to other data points

Low Variance - Results are grouped together

High Variance - Results are far apart from each other

# Linear Regression

Trying to predict the next value in a continuum

Quantitative Prediction

Housing attributes to get a House Price

- Inputs: number of rooms, sq. meters, neighborhood median salary, age of the home (historical data)

- Output: Price of the home

https://scikit-learn.org/0.15/modules/linear_model.html

###### Pros

Linear Regression is simple to implement.

Less complexity compared to other algorithms.

###### Cons

Outliers affect this algorithm badly.

It over-simplifies real-world problems by assuming a linear relationship among the variables

Not recommended for practical use-cases.

###### Mean Absolute Error (MAE)
Average of the absolute value of the error differentials

How far off was the model fro the actual

Won't punish large errors so we could have many outliers over and under but the model would look like it fits

Think Standard Deviation but for Machine Learning

###### Mean Squared Error (MSE)
Solves the problem of outliers as well

But squares the outcome (Dollars Squared, doesn't make sense)

###### Root Mean Squared Error (RMSE)
To fix the issue of massive outliers and data type squared, we use Root Mean Squared and square root the MSE

This is the most reliable but also knowing the data helps as well

`
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions)) # Ideal because is shows how off we are per prediction
`

```python


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# SKLearn Algorithm Imports

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

```

```python
# Get the dataset and create a dataframe from the dataset


df = pd.read_csv('USA_Housing.csv')


```

```python
# Train test and split of the data
# First split the data into X and y
# X is the attributes
# y is the target or in this case Price

X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

y = df['Price']



```

```python
# Train Test Split
# Anything outside of learning environments, DO NOT USE random state
# We want RANDOMNESS in our machine learning

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# Now we begin training our model
# We train our model on X_train, y_train
# We do this because we want to allow the model to find the line of best fit for our data

# Instantiate the model
model = LinearRegression()

```

```python
# Fit or Train the model

model.fit(X_train, y_train)
```

```python
# Now that the model is done training, we can send it predictions in the real world
# ...but before doing that, we might want to see how our model performs with test data

predictions = model.predict(X_test)


```

```python
# Can we do automated evaluation
# YES!!


# Mean Absolute Error
print('MAE', mean_absolute_error(y_test, predictions))

# Mean Squared Error
print('MSE', mean_squared_error(y_test, predictions))
# $^2


# Root Mean Squared Error
print('RMSE', np.sqrt(mean_squared_error(y_test, predictions)))
# $

# Score
print('Score', model.score(X_test, y_test))
```
